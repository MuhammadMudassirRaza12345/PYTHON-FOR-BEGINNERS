
create a bank class that will do all banking-related tasks. you will create one class with all methods related to banking. This class holds two values. one is the total money amount for the user and user name.

Once any object for this class is created, it will call the init(self) method, which will also be known as the constructor. Inside this method, you will be setting the total amount as 0 and the user names as an empty string.

In this class you will have four methods:
1. welcome, 2. printcurrentbalance , 3. deposit, 4. withdraw

"welcome" will used to greet the user and it also records the user name and keep it in the name variable.

"printcurrentbalance" will be used to print the current balance i.e. total_amount
deposit method is used to deposit money. It takes the amount from the user and adds it to the total_amount.

The "withdraw" method will be used to withdraw money. It will take the amount and compares it with total_amount. If the amount will greater than total_amount, it will show one insufficient balance message. Else, it will decrements that amount from total_amount.

when you will run this program, it will create one Bank object in the beginning, which will be the bank. It calls welcome method to show a greeting message and records the name.

It will run one infinite loop. It will keep taking user inputs and based on the value, it will call different methods of a bank.
